```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1a = np.load('Conc_1_Angle_0.00_Type_reference_Dimension_2.odor_current_input.npy')
a2a = np.load('Conc_0.8_Angle_0.00_Type_reference_Dimension_2.odor_current_input.npy')
ba = np.load('Conc_1_Angle_30.00_Type_30_Dimension_2.odor_current_input.npy')
ca = np.load('Conc_1_Angle_60.00_Type_60_Dimension_2.odor_current_input.npy')
da = np.load('Conc_1_Angle_90.00_Type_90_Dimension_2.odor_current_input.npy')
ea = np.load('Conc_1_Angle_180.00_Type_180_Dimension_2.odor_current_input.npy')
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import LocallyLinearEmbedding
```

```python
model = PCA(n_components=3)
```

```python

model.fit(ba.T)
model.fit(ca.T)
model.fit(da.T)
#model.fit(c.T)
model.fit(a1a.T)
#model.fit(a2.T)

a1 = model.transform(a1a.T)
a2 = model.transform(a2a.T)
b = model.transform(ba.T)
c = model.transform(ca.T)
d = model.transform(da.T)
e = model.transform(ea.T)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure()
fig.patch.set_alpha(0.0)
ax = fig.add_subplot(111, projection='3d')
ax.patch.set_alpha(0.0)
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.scatter(a1[::5,0],a1[::5,1],a1[::5,2],label='Odor A1',alpha=0.2,s=2)
ax.scatter(a2[::5,0],a2[::5,1],a2[::5,2],label='Odor A2',alpha=0.2,s=2)
ax.scatter(d[::5,0],d[::5,1],d[::5,2],label='Odor B',alpha=0.2,s=2)
#ax.scatter(c[::5,0],c[::5,1],c[::5,2],label='Odor C',alpha=0.2,s=2)
ax.scatter(e[::5,0],e[::5,1],e[::5,2],label='Odor C',alpha=0.2,s=2)
ax.grid(False)
plt.axis('off')
plt.legend(frameon=False)
plt.show()
```
