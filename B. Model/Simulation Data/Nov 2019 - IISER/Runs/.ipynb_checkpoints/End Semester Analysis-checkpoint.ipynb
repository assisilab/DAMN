```python
def runningResponsive(x,N):
    y = np.zeros(x.shape[1])
    for ctr in range(int(N/2),x.shape[1]-int(N/2)):
         y[ctr] = (np.sum(x[:,ctr-int(N/2):ctr+int(N/2)],axis=1)>0).mean()
    return y
```
