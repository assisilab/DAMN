```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
temp = [] 
files = list(filter(lambda v:"1.npy" in v,os.listdir()))
files.sort()
for i in files:
    temp.append(np.load(i)[1:,:])
```

```python
plt.imshow(np.concatenate(temp)[::100,:90][500:2500,:].T,aspect='auto')
```

```python
np.save("values",np.concatenate(temp)[::100,:90][500:2500,:].T)
```
