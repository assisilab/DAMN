```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
duration = 60
resolution =0.01

n_KC = 15000
n_PN = 90

pn_spikes = np.random.choice([0,1],size=(int(duration/resolution),n_PN),p=[0.999,0.001])
```

```python
pn_spikes_dec = np.int32(np.sum(np.vsplit(pn_spikes,int(pn_spikes.shape[0]*resolution)),axis=1)>0)
```

```python
cmatrix = np.random.choice([0,1],size=(n_KC,n_PN))
```

```python
u = 0.0012 + np.random.uniform(-0.00068,0.00068,n_KC)
Be = 0.03
sigma = 0.06 + np.random.exponential(0.0072,n_KC)
sigmae = 1
alpha= 3.65

x_KC = (sigma-1)*np.ones((int(duration)+2,n_KC))
y_KC = x_KC - alpha/(1-x_KC)

# I_KC = 0.05*np.ones((int(duration),n_KC))

n_syn = np.sum(cmatrix)
g_PNKC = np.zeros((int(duration)+2,n_syn))
I_KC = np.zeros((int(duration)+2,n_KC))
gamma = 0.4
G_PNKC = 0.00066
S_den = 165e-6

E_ach = 0

def f_alpha(xn,xnm1,u):
    return np.where(xn <=0, alpha/(1-xn)+ u, np.where(np.logical_and(xnm1<=0,xn<alpha+u),alpha+u,-np.ones(n_KC)) )

for i in range(2,x_KC.shape[0]):
    print("{}/{}".format(i,x_KC.shape[0]),end='\r')
    
    g_PNKC[i,:] = gamma*g_PNKC[i-1,:] + G_PNKC/S_den * (cmatrix*pn_spikes_dec[i-2,:])[cmatrix>0]
    
    I_KC_t = np.float64(np.copy(cmatrix))
    np.putmask(I_KC_t,cmatrix>0,g_PNKC[i-1,:])
    I_KC[i,:] = (-I_KC_t.T*(x_KC[i,:]-E_ach)).sum(axis=0)
    
    x_KC[i,:] = f_alpha(x_KC[i-1,:],x_KC[i-2,:],y_KC[i-1,:]+Be*I_KC[i,:])
    y_KC[i,:] = y_KC[i-1,:] - u*(1+x_KC[i-1,:]) + u*(sigma+sigmae*I_KC[i,:])
```
